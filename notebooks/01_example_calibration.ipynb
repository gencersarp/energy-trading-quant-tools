# Model Calibration Example
Loading synthetic data and fitting a seasonal trend model.

```
import numpy as np
from energytrading.data.synthetic import generate_synthetic_power_data
from energytrading.models.seasonal import SeasonalTrendModel

df = generate_synthetic_power_data(days=100)
t = np.arange(len(df))
model = SeasonalTrendModel(fourier_order=5)
model.fit(t, df['price'].values)
print('Model fitted successfully!')
```
